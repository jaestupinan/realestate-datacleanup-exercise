#### Exercise 18. Normalize the variable of prices for each population and plot the 4 histograms in the same plot (you must use the subset obtained in Exercise 14) (★★★)

For the normalization method, you can use the one you consider; there is not a single correct answer to this question. Print the plot and write in the Markdown cell a brief analysis about the plot.

Hint: You can help yourself by reviewing the *multihist* demo of Matplotlib.

```python
# Normalizar precios para cada población
import matplotlib.pyplot as plt

# Definir la función de normalización Min-Max
def normalize_minmax(data):
    return (data - data.min()) / (data.max() - data.min())

# Normalizar los precios por grupo de población
for population in south_belt_population:south_belt.loc[south_belt['level5'] == population, 'price_normalized'] = normalize_minmax(south_belt.loc[south_belt['level5'] == population, 'price'])

# Crear el gráfico
fig, ax = plt.subplots()

# Graficar histogramas para cada población
for population in south_belt_population:
    ax.hist(south_belt.loc[south_belt['level5'] == population, 'price_normalized'],alpha=0.5,label=population)

# Añadir etiquetas y título
ax.set_xlabel('Precio normalizado')
ax.set_ylabel('Frecuencia')
ax.set_title('Distribución de precios normalizados en el cinturón sur')
ax.legend()

# Mostrar el gráfico
plt.show()
```

#### Exercise 19. What can you say about the price per square meter (price/m2) between the towns of "Getafe" and "Alcorcón"? You must use the subset obtained in Exercise 14 (★★☆)

Hint: Create a new column called `pps` (price per square meter) and then analyze the values.

```python
import numpy as np
import matplotlib.pyplot as plt

# Crear 'pps' (precio por metro cuadrado)
south_belt['pps'] = south_belt['price'] / south_belt['surface'].replace(0, np.nan)  # Evitar división por cero

# Filtrar las ciudades de Getafe y Alcorcón
getafe_alcorcon = south_belt[south_belt['level5'].isin(['Getafe', 'Alcorcón'])]

# Verificar si hay datos para Getafe y Alcorcón
if not getafe_alcorcon.empty:
    # Análisis de pps
    print(getafe_alcorcon.groupby('level5')['pps'].describe())
    
    # Extraer valores para el boxplot
    getafe_pps = getafe_alcorcon.loc[getafe_alcorcon['level5'] == 'Getafe', 'pps'].values
    alcorcon_pps = getafe_alcorcon.loc[getafe_alcorcon['level5'] == 'Alcorcón', 'pps'].values
    
    # Eliminar NaN (si los hay)
    getafe_pps = getafe_pps[~np.isnan(getafe_pps)]
    alcorcon_pps = alcorcon_pps[~np.isnan(alcorcon_pps)]
    
    # Verificar si hay datos suficientes para graficar
    if len(getafe_pps) == 0 or len(alcorcon_pps) == 0:
        print("No hay datos suficientes para graficar.")
    else:
        # Crear el boxplot
        fig, ax = plt.subplots()
        ax.boxplot([getafe_pps, alcorcon_pps], labels=['Getafe', 'Alcorcón'])
        plt.title('Precio por metro cuadrado en Getafe y Alcorcón')
        plt.xlabel('Municipio')
        plt.ylabel('Precio por metro cuadrado (€/m²)')
        plt.show(block=True)  # Forzar la visualización
else:
    print("No hay datos para Getafe o Alcorcón.")
```

```python
from ipyleaflet import Map, basemaps, CircleMarker, LayerGroup
import pandas as pd

# Crear el mapa centrado en Madrid
map = Map(center=(40.3, -3.7), zoom=10, basemap=basemaps.OpenStreetMap.Mapnik)

# Crear una lista para almacenar los marcadores
markers = []

# Definir colores para cada población
colors = {
    'Getafe': 'red',
    'Alcorcón': 'blue',
    'Leganés': 'green',
    'Fuenlabrada': 'yellow'
    }

# Iterar sobre cada fila del DataFrame south_belt
for index, row in south_belt.iterrows():
    # Crear un marcador para cada coordenada
    marker = CircleMarker(
        location=(row['latitude'], row['longitude']),  # Coordenadas
        radius=5,  # Tamaño del marcador
        color=colors.get(row['level5'], 'black'),  # Color según la población
        fill=True  # Rellenar el marcador
    )
    markers.append(marker)  # Agregar el marcador a la lista

# Crear un grupo de capas para los marcadores
layer_group = LayerGroup(layers=markers)

# Agregar el grupo de capas al mapa
map.add_layer(layer_group)

# Mostrar el mapa
map
```
